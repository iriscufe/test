## 2.2 BeautifulSoup
通俗来说是【解析、遍历、维护“标签树”(例如html、xml等格式的数据对象)的功能库 】

BeautifulSoup类可以通过以下基本元素去提取html中的内容：

- Tag，以<>开头和</>结尾的标签，例：obj.Tag1.Tag2
- Name，标签内容，例：obj.Tag1.name
- Attributes，标签的属性，例：obj.Tag1.attrs
- NavigableString,标签里的字符串，例：obj.Tag1.string
- Comment，标签内字符串注释的部分，例：obj.Tag1.string

#### 2.2.1 知道需要爬的标签所有信息


````python
from bs4 import BeautifulSoup
url="http://www.pythonscraping.com/pages/warandpeace.html"
html=urlopen(url)
bsObj=BeautifulSoup(html.read(),"html.parser")####html.parser是内置的BeautifulSoup解析器
print("1、通过Tag爬取：\n"+str(bsObj.html.body.span))
print("2、通过name爬取：\n"+str(bsObj.html.span.name))
print("3、通过Attributes爬取：\n"+str(bsObj.html.span.attrs))
print("4、通过NavigableString爬取：\n"+str(bsObj.html.span.string))
'''
注意：用对象引用下一级对象的方式只能显示它第一个解析出的标签内容
'''
````

#### 2.2.1 只知道上级和下级节点信息

- 子标签（children）是父标签的下一级
- 后代标签（descendants）是父标签下面所有级别的标签
- 兄弟标签（siblings）是同级的所有标签
- 父标签（parent）是上一级标签

````python
from urllib.request import urlopen
from bs4 import BeautifulSoup
html=urlopen("http://www.pythonscraping.com/pages/page3.html")
bsObj=BeautifulSoup(html,"html.parser")
print("子标签内容：")
for child in bsObj.find("table",{"id":"giftList"}).children: # descendants
    print(child)
print("兄弟标签内容：")
for sibling in bsObj.find("table",{"id":"giftList"}).tr.next_siblings: # previous_siblings
    print(sibling)
print("父标签内容：")
print(bsObj.find("img",{"src":"../img/gifts/img1.jpg"}).parent.previous_sibling.get_text())
````

#### 2.2.1 爬取符合条件数据

采用正则表达式来采集更多条件的数据

- runoo+b，可以匹配 runoob、runooob、runoooooob 等，+ 号代表前面的字符必须至少出现一次（1次或多次）
- runoo\*b，可以匹配 runob、runoob、runoooooob 等，* 号代表字符可以不出现，也可以出现一次或者多次（0次、或1次、或多次）
- colou?r 可以匹配 color 或者 colour，? 问号代表前面的字符最多只可以出现一次（0次、或1次）

**正则表达式传送门：https://www.runoob.com/regexp/regexp-syntax.html**

````python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
html=urlopen("http://www.pythonscraping.com/pages/page3.html")
bsObj=BeautifulSoup(html,"html.parser")
images=bsObj.findAll("img",{"src":re.compile("\.\./img/gifts/img.\.jpg")})
for image in images:
    print(image.attrs["src"])
````

#### 2.2.1 网页无法连接
打不开网页，程序将报错，执行不下去。
```
try:
    code #运行别的代码
except <name>： 
    <语句> #如果在try部份引发了'name'异常
except <name>，<数据>:
    <语句> #如果引发了'name'异常，获得附加的数据
else:
    <语句> #如果没有异常发生
```
**Python错误类型定义传送门：https://www.runoob.com/python/python-exceptions.html**

````python
from urllib.error import URLError
from urllib.request import urlopen
from bs4 import BeautifulSoup
def get_title(url):
    try:
        html=urlopen(url)
        bsObj=BeautifulSoup(html.read(),"html.parser")
        return bsObj.html.body.h1
    except :
        print("url连接错误")
title=get_title("https://www.wikipedia.org/")
print(title)
````

#### 2.2.1 爬取内容跨页

````python
from urllib.error import URLError
from urllib.request import urlopen
from bs4 import BeautifulSoup
url = "https://movie.douban.com/top250"
html=urlopen(url)
bsObj=BeautifulSoup(html.read(),"html.parser")
movie_list = bsObj.find("li")
nList = len(movie_list)
for i in range(0,250,nList):
    url = "https://movie.douban.com/top250?start="+str(i)
    html=urlopen(url)
    bsObj=BeautifulSoup(html.read(),"html.parser")
    movie_list = bsObj.findAll("span",text=re.compile("^[\u4e00-\u9fa5]{0,}$"))
    for j in movie_list:
        print(j.get_text())
````
